```python
#Basit Bir Çizgi Grafiği Çizmek:

#pyplot modülü: çizgiler ve grafikler oluşturan fonksiyonlar içerir.
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]

# subplots: aynı şekil üzerinde bir veya daha fazla grafik üretebilir.
# fig değişkeni, şeklin tamamını veya üretilen grafikler koleksiyonunu temsil eder.
# ax değişkeni ise: şekildeki tek bir grafiği temsil eder.
fig, ax = plt.subplots()

#plot(): kendisine verilen veriyi anlamlı bir şekilde çizmeye çalışır.
ax.plot(squares)

# Matplotlib görüntüleyicisini açar ve grafikte gezinmemizi sağlar.
plt.show()
```

```python
#Etiket Tipini ve Çizgi Klaınlığını Değiştirmek:
#Grafik okunurluğunu arttırmak için yaparız.

import matplotlib.pyplot as plt
squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(squares, linewidth=3)

# Çizelge başlığını belirle ve eksenlere etiket ver.
# Çizelge Başlığı
ax.set_title("Sayıların Karesi", fontsize=24)
# Eksenlere etiket veriyoruz.
ax.set_xlabel("Değer", fontsize=14)
ax.set_ylabel("Değerin Karesi, fontsize=14")

# Çentik etiketlerinin büyüklüğünü belirle.
ax.tick_params(axis="both", labelsize=14)

plt.show()
```

```python
#plot() ' a bir sayı dizisi verdiğimizde ilk veri noktasının değeri 0 olan bir x koordinatına karşılık geldiğini varsayar.

#Ancak bizim grafiğimizin ilk noktasının değeri 1 olan x' e karşılık gelmektedir.
# plot()' a girdi, çıktı değerleri vererek bu durumu geçersiz kılabiliriz.

import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth=5)

ax.set_title("Sayıların Karesi", fontsize=15)
ax.set_xlabel("Değer", fontsize=15)
ax.set_ylabel("Değerin Karesi", fontsize=15)

ax.tick_params(axis="both", labelsize=15)

plt.show()
```

```python
#RandomWalk Sınıfı' nı Oluşturmak:
#RandomWalk Sınıf için 2 metoda ihtiyacımız olacak: __init__() ve yürüyüşteki noktaları hesaplayacak, fill_walk().

import matplotlib.pyplot as plt
from random_walk import RandomWalk

"""Yürüyüşteki bütün çizgileri çizdiren kod"""
# Rastgele bir yürüyüş yap.
rw = RandomWalk()
rw.fill_walk()
# Yürüyüşteki noktaları çiz.
plt.style.use("classic")
fix, ax = plt.subplots()
ax.scatter(rw.x_values, rw.y_values, c="red", s=200)

plt.show()
```

```python
# Birden Çok Ratgele Yürüyüş Üretmek:
import matplotlib.pyplot as plt
from random_walk import RandomWalk

# Program aktif olduğu sürece yeni yürüyüş yapmayı sürdür.
while True:
    rw = RandomWalk()
    rw.fill_walk()
    
    plt.style.use("classic")
    fix, ax = plt.subplots()
    ax.scatter(rw.x_values, rw.y_values, c=rw.y_values, cmap=plt.cm.Blues, s=500)
    
    plt.show()
    
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == "n":
        break
```

```python
#Yürüyüşe Stil Vermek:
# Noktaları Renklendirmek:
# Yürüyüşteki noktaların sıralamasını göstermek için  bir renk haritası kullanıyoruz. 
# Daha sonra noktaların rengi daha açık olsun diye herbir noktadan siyah çerçeveyi kaldırıyoruz.
# Yürüyüşteki konumlarına göre noktaları renklendirmek için c argümanına her bir noktanın konumunu içeren  bir liste gönderdik.
#Noktalar sırayla çizildiği için liste aşağıda gösterildiği gibi 0' dan 4999' a kadar olan sayıları içerir.

import matplotlib.pyplot as plt
from random_walk import RandomWalk

"""Yürüyüşteki bütün çizgileri çizdiren kod"""
while True:
    # Rastgele bir yürüyüş yap.
    rw = RandomWalk()
    rw.fill_walk()
    # Yürüyüşteki noktaları çiz.
    plt.style.use("classic")
    fix, ax = plt.subplots()
    point_numbers = range(rw.num_points)
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap= plt.cm.Blues, edgecolors="none", s=300)
    ax.scatter(rw.x_values, rw.y_values, s=300)
    
    plt.show()
    
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == "n":
        break
```

```python
#Başlangıç ve Bitiş Noktalarını Çizdirmek:
#Yürüyüş boyunca olan konumlarını göstermek için noktaları renklendirmeye ek olarak her bir yürüyüşün nerede başladığını ve nerede bittiğini görmek faydalı olacaktır.
# Bunu yapmak için, ana seri çizildikten sonra ilk ve son noktaları tek tek çizeriz.

import matplotlib.pyplot as plt
from random_walk import RandomWalk

"""Yürüyüşteki bütün çizgileri çizdiren kod"""
while True:
    # Rastgele bir yürüyüş yap.
    rw = RandomWalk()
    rw.fill_walk()
    # Yürüyüşteki noktaları çiz.
    plt.style.use("classic")
    fix, ax = plt.subplots()
    point_numbers = range(rw.num_points)
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap= plt.cm.Blues, edgecolors="none", s=500)
    #İlk ve son noktaları vurgula.
    ax.scatter(0, 0, c="green", edgecolors="none", s=500)
    ax.scatter(rw.x_values[-1], rw.y_values[-1], c="red", edgecolors="none", s=500)
    
    plt.show()
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == "n":
        break
```

```python
#Eksenleri Temizlemek:
# Her bir yürüyüşün yoluyla karışmaması için bu grafikteki eksenleri kaldırıyoruz.

import matplotlib.pyplot as plt
from random_walk import RandomWalk

"""Yürüyüşteki bütün çizgileri çizdiren kod"""
while True:
    # Rastgele bir yürüyüş yap.
    rw = RandomWalk()
    rw.fill_walk()
    # Yürüyüşteki noktaları çiz.
    plt.style.use("classic")
    fix, ax = plt.subplots()
    point_numbers = range(rw.num_points)
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap= plt.cm.Blues, edgecolors="none", s=500)
    #İlk ve son noktaları vurgula.
    ax.scatter(0, 0, c="green", edgecolors="none", s=500)
    ax.scatter(rw.x_values[-1], rw.y_values[-1], c="red", edgecolors="none", s=500)
    
    #Eksenleri Kaldır
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
    plt.show()
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == "n":
        break
```
